```python
import xarray as xr
import numpy as np
import zipfile
import matplotlib.pyplot as plt
    
temperature_data = xr.open_dataset("../data/climate/extracted/temperature.nc")["t2m"]

time_coord = temperature_data.coords["valid_time"]


combined_data = xr.merge([temperature_data])
```

```python
selection = combined_data.sel(valid_time="2019-01-01",
                              longitude = slice(5,22),
                              latitude =  slice(54, 38))["t2m"].plot()
plt.show()
```

```python
avg = combined_data["t2m"].mean(dim=("longitude"))
```

```python
avg.sel(valid_time = slice("2019-01-01", "2020-01-01")).plot(x = "valid_time", y = "latitude")
plt.show()
```
